### Importa√ß√£o dos dados



```python
import pandas as pd

# Carregar os dados
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Exibir informa√ß√µes sobre o DataFrame
loja1.info()

# Exibir as 5 primeiras e as 5 √∫ltimas linhas do DataFrame
loja1
```

# 1. An√°lise do faturamento


```python
import numpy as np

# Calculando o faturamento total por loja
faturamento1 = loja1["Pre√ßo"].sum()
faturamento2 = loja2["Pre√ßo"].sum()
faturamento3 = loja3["Pre√ßo"].sum()
faturamento4 = loja4["Pre√ßo"].sum()

# Exibindo os resultados
print("üìä Faturamento total por loja:\n")
print(f"Loja 1: R$ {faturamento1:,.2f}")
print(f"Loja 2: R$ {faturamento2:,.2f}")
print(f"Loja 3: R$ {faturamento3:,.2f}")
print(f"Loja 4: R$ {faturamento4:,.2f}")
```

```python
import matplotlib.pyplot as plt

# Dados de exemplo
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
valores = [faturamento1, faturamento2, faturamento3, faturamento4]
cores = ["blue", "green", "purple", "orange"]

# Plot
plt.figure(figsize=(10, 6))
barras = plt.barh(lojas, valores, color=cores)

plt.title("Faturamento Total por Loja", fontsize=16)
plt.xlabel("Faturamento (R$ milh√µes)", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.5)

# Adiciona os valores dentro das barras
for barra in barras:
    plt.text(barra.get_width() * 0.98,
             barra.get_y() + barra.get_height() / 2,
             f'R$ {barra.get_width():,.0f}',
             va='center', ha='right', fontsize=11, color='white', weight='bold')

plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
import seaborn as sns

# Construindo tabela cruzada: categorias x lojas
categorias = sorted(set(loja1["Categoria do Produto"]) |
                    set(loja2["Categoria do Produto"]) |
                    set(loja3["Categoria do Produto"]) |
                    set(loja4["Categoria do Produto"]))

dados = {
    'Loja 1': loja1['Categoria do Produto'].value_counts(),
    'Loja 2': loja2['Categoria do Produto'].value_counts(),
    'Loja 3': loja3['Categoria do Produto'].value_counts(),
    'Loja 4': loja4['Categoria do Produto'].value_counts(),
}

df_heatmap = pd.DataFrame(dados).fillna(0).reindex(categorias)

plt.figure(figsize=(10, 6))
sns.heatmap(df_heatmap, annot=True, fmt='.0f', cmap="coolwarm",)
plt.title("Quantidade de Vendas por Categoria e Loja", fontsize=14)
plt.ylabel("Categoria")
plt.xlabel("Loja")
plt.show()
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt

lojas = [loja1, loja2, loja3, loja4]
titulos = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
cores = ["blue", "green", "purple", "orange"]  # Cores definidas para cada loja

# Estilo do seaborn
sns.set(style="whitegrid")

# Criar os subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

for i in range(4):
    notas = lojas[i]["Avalia√ß√£o da compra"]
    media = notas.mean()

    sns.histplot(notas, bins=5, kde=True, ax=axs[i], color=cores[i], edgecolor='black')
    axs[i].axvline(media, color='darkred', linestyle='--', linewidth=2, label='M√©dia')
    axs[i].text(media + 0.1, axs[i].get_ylim()[1] * 0.9,
                f"¬µ = {media:.2f}", color='darkgreen', fontsize=10)

    axs[i].set_title(titulos[i])
    axs[i].set_xlim(0, 5)
    axs[i].set_xlabel("Avalia√ß√£o")
    axs[i].set_ylabel("Frequ√™ncia")
    axs[i].legend()

# Ajustes finais
fig.suptitle("Distribui√ß√£o das Avalia√ß√µes por Loja", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
from wordcloud import WordCloud

lojas = [loja1, loja2, loja3, loja4]
titulos = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
cores = ['Blues', 'Greens', 'Purples', 'Oranges']  # 1 cor diferente por loja

# Criar gr√°fico 2x2 com as nuvens
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

for i in range(4):
    contagem_produtos = lojas[i]['Produto'].value_counts()
    wordcloud = WordCloud(width=800, height=400,
                          background_color='white',
                          colormap=cores[i],
                          max_words=100).generate_from_frequencies(contagem_produtos)

    axs[i].imshow(wordcloud, interpolation='bilinear')
    axs[i].set_title(titulos[i], fontsize=14)
    axs[i].axis('off')

fig.suptitle("Nuvem de Palavras - 100 Produtos Mais Vendidos por Loja", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# import pandas as pd
# import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.patches import Patch

# # Recarregar os arquivos CSV
# loja1 = pd.read_csv("/mnt/data/loja_1.csv")
# loja2 = pd.read_csv("/mnt/data/loja_2.csv")
# loja3 = pd.read_csv("/mnt/data/loja_3.csv")
# loja4 = pd.read_csv("/mnt/data/loja_4.csv")

# Listas auxiliares
lojas = [loja1, loja2, loja3, loja4]
titulos = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
colormaps = ["Blues", "Greens", "Purples", "Oranges"]

# Criar subplots 2x2
fig, axs = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Top 5 Mais e Menos Vendidos por Loja", fontsize=18)

for i, ax in enumerate(axs.flat):
    df = lojas[i]
    titulo = titulos[i]
    cmap_name = colormaps[i]

    contagem = df["Produto"].value_counts()

    # Top e Bottom 5 em ordem decrescente
    top5 = contagem.head(5)
    bottom5 = contagem.tail(5).sort_values(ascending=False)
    juntos = pd.concat([top5, bottom5])

    # Normalizar para intensidade da cor
    norm = mcolors.Normalize(vmin=juntos.min(), vmax=juntos.max())
    cmap = plt.get_cmap(cmap_name)
    cores = [cmap(norm(v)) for v in juntos.values]

    # Desenhar barras com borda
    bars = ax.barh(juntos.index, juntos.values, color=cores, edgecolor='black')
    ax.set_title(titulo, fontsize=14)
    ax.set_xlabel("Quantidade Vendida")
    ax.invert_yaxis()

    # Linha tracejada separando top5 e bottom5
    ax.axhline(y=4.5, color='gray', linestyle='--', linewidth=1.2)

    # Adicionar legenda
    legenda = [
        Patch(facecolor=cmap(norm(top5.max())), edgecolor='black', label='Top 5 Mais Vendidos'),
        Patch(facecolor=cmap(norm(bottom5.max())), edgecolor='black', label='Top 5 Menos Vendidos')
    ]
    ax.legend(handles=legenda, loc='lower right')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import geopandas as gpd

# Agrupar por estado e contar vendas
estados_loja1 = loja1["Local da compra"].value_counts()
estados_loja2 = loja2["Local da compra"].value_counts()
estados_loja3 = loja3["Local da compra"].value_counts()
estados_loja4 = loja4["Local da compra"].value_counts()

# Baixar mapa dos estados do Brasil via GeoPandas
br = gpd.read_file("https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/brazil-states.geojson")

# Ajustar nomes dos estados para siglas no DataFrame
br["sigla"] = br["name"].map({
    'Acre': 'AC', 'Alagoas': 'AL', 'Amap√°': 'AP', 'Amazonas': 'AM',
    'Bahia': 'BA', 'Cear√°': 'CE', 'Distrito Federal': 'DF', 'Esp√≠rito Santo': 'ES',
    'Goi√°s': 'GO', 'Maranh√£o': 'MA', 'Mato Grosso': 'MT', 'Mato Grosso do Sul': 'MS',
    'Minas Gerais': 'MG', 'Par√°': 'PA', 'Para√≠ba': 'PB', 'Paran√°': 'PR',
    'Pernambuco': 'PE', 'Piau√≠': 'PI', 'Rio de Janeiro': 'RJ', 'Rio Grande do Norte': 'RN',
    'Rio Grande do Sul': 'RS', 'Rond√¥nia': 'RO', 'Roraima': 'RR', 'Santa Catarina': 'SC',
    'S√£o Paulo': 'SP', 'Sergipe': 'SE', 'Tocantins': 'TO'
})

# Adicionar colunas de vendas por estado no GeoDataFrame
for loja_nome, vendas in zip(["Loja 1", "Loja 2", "Loja 3", "Loja 4"],
                              [estados_loja1, estados_loja2, estados_loja3, estados_loja4]):
    br[loja_nome] = br["sigla"].map(vendas).fillna(0)

# Plotar os 4 mapas lado a lado
fig, axs = plt.subplots(2, 2, figsize=(16, 12))
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
cores = ["Blues", "Greens", "Purples", "Oranges"]

for i, ax in enumerate(axs.flat):
    loja = lojas[i]
    br.plot(column=loja, cmap=cores[i], linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
    ax.set_title(f'Densidade de Vendas - {loja}', fontsize=14)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Carregar arquivos CSV
lojas = [loja1, loja2, loja3, loja4]
titulos = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
cores = ["blue", "green", "purple", "orange"]

# Mapa do Brasil
br = gpd.read_file("https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/brazil-states.geojson")

# Criar gr√°fico 2x2
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

for i, ax in enumerate(axs.flat):
    df = lojas[i]
    
    # Agrupar por estado
    agrupado = df.groupby("Local da compra").agg({
        "lat": "mean",
        "lon": "mean",
        "Produto": "count"
    }).rename(columns={"Produto": "quantidade"}).reset_index()
    
    # Plotar mapa base
    br.plot(ax=ax, color="whitesmoke", edgecolor="gray")
    
    # Plotar bolinhas proporcionais
    ax.scatter(
        agrupado["lon"],
        agrupado["lat"],
        s=agrupado["quantidade"] * 2,  # ajuste o fator de escala se necess√°rio
        color=cores[i],
        alpha=0.6,
        label='Quantidade de vendas'
    )
    
    ax.set_title(f"Concentra√ß√£o de Vendas - {titulos[i]}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
import folium
from folium.plugins import HeatMap

def criar_mapa_loja(df, nome_loja, cor='blue'):
    mapa = folium.Map(location=[-15.78, -47.93], zoom_start=4, tiles="CartoDB positron")
    
    for _, row in df.iterrows():
        folium.CircleMarker(
            location=[row['lat'], row['lon']],
            radius=3,
            color=cor,
            fill=True,
            fill_opacity=0.6
        ).add_to(mapa)
    
    mapa.save(f"mapa_{nome_loja.lower().replace(' ', '_')}.html")
    return mapa

# Exemplo de uso:
mapa1 = criar_mapa_loja(loja1, "Loja 1", "blue")
mapa2 = criar_mapa_loja(loja2, "Loja 2", "green")
mapa3 = criar_mapa_loja(loja3, "Loja 3", "purple")
mapa4 = criar_mapa_loja(loja4, "Loja 4", "orange")

```
